```python
### Mind wandering labels ###
#dependecies
import pandas as pd
```

```python
# paths
read_path_labels = r"W:\WCT\04_Mind-Wandering-Labstudy\04_Daten\04_Prepared_data\01_MW_Probes\MW_Probes.csv"
read_path_features = r"W:\WCT\04_Mind-Wandering-Labstudy\04_Daten\04_Prepared_data\00_Julia\features_final.csv"
save_path = r"W:\WCT\04_Mind-Wandering-Labstudy\04_Daten\04_Prepared_data\00_Julia\Model Building\features_with_label_awareness.csv"

# read Data
df_labels = pd.read_csv(read_path_labels)
df_features = pd.read_csv(read_path_features)
df_labels.head()

df_labels.replace("P04","4", inplace= True)
```

```python
#Inofs:
# we find the infos we need in the column Wareness_cat und Awareness_cat_test

# Kategorien:
# im Falle von MW:  Task-Unrelated-Thought (TUT) == Mind Wandering 
# aware: they are aware that mind wandering was happen
# unaware: thaey were not aware that mind wandering was happen
# ['OnTask/TRT/TRI']: alle Kategorien die kein Mind Wandering abbilden

# numbers as categories
# ['OnTask/TRT/TRI', 'aware', 'unaware']
# 3	= OnTask/TRT/TRI
# 2 = aware
# 1 = unaware


# Probe = Gedankenstichprobe
# Subjectname = Participants

# we want to match the Data with probe number and subjectnumber to our features
#Du kannst sie über den Personenidentifier und die Probe Nummer zu den Eye Tracking Daten matchen.
```

```python
df_labels["Awareness"].unique()
```

```python
# exclude the participants, that we have the same partcicpiant sin labels and features
excluded_participants = [ str(i) for i in [2, 8, 11, 12, 13, 24, 25, 27, 30, 31, 38, 41, 43, 47, 50, 52, 54, 59, 63, 69, 70, 75, 84, 89, 91, 101, 102, 103, 104, 105, 999]]
df_labels = df_labels.loc[~df_labels["SubjectName"].isin(excluded_participants)]
df_labels["SubjectName"].unique()
```

```python
# make the dataframes more similar, that we can merge
df_features["Stimulus"] = df_features["Stimulus"].str.replace("vid", " ").astype("int")

df_labels = df_labels.rename(columns={"SubjectName": "Participant"})
df_features = df_features.rename(columns = {"Stimulus": "Probe"})
df_labels["Participant"] = df_labels["Participant"].astype("int")
df_features["Participant"] = df_features["Participant"].astype("int")
```

```python
df_features
```

```python
# we want to merge only the necessary columns
df_labels_necessary = df_labels[["Participant", "Probe", "Awareness_cat"]]
```

```python
# merging and saving as csv
df_features_with_label = pd.merge(df_features, df_labels_necessary, on = ["Participant","Probe"])
df_features_with_label.to_csv(save_path)
```
