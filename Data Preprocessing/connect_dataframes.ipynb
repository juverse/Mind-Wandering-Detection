```python
#################### code with big dataset #####################
# sort list 
participants = df_right["Participant"].unique()
list_participants = (list(participants) *15)
list_participants_sorted = [int(x) for x in participants] *15
list_participants_sorted.sort()
#h *15
list_sorted = [str(x) for x in list_participants_sorted]
```

```python
# dataframe with all possible participants and videos
data_all_videos_partipants = {  "Stimulus": ["vid1","vid10","vid11","vid12","vid13","vid14","vid15","vid2","vid3","vid4","vid5","vid6","vid7","vid8","vid9"] * 71,
    "Participant": list_sorted
   }

df_all_videos_partipants = pd.DataFrame(data_all_videos_partipants)
df_all_videos_partipants
```

```python
df_left_all_rows = pd.merge(df_all_videos_partipants,df_left, how = "left")
df_left_all_rows
```

```python
df_right_all_rows = pd.merge(df_all_videos_partipants,df_right, how = "left")
df_right_all_rows
```

```python
df_vergence
df_vergence_all_rows = pd.merge(df_all_videos_partipants,df_vergence, how = "left")
df_vergence_all_rows
```

```python
df = pd.concat([df_left_all_rows, df_right_all_rows]).mean(level = 0)
df.insert(1, "Stimulus", df_left_all_rows["Stimulus"], True)
#df["Stimulus"] = df_left_all_rows["Stimulus"]
df["Participant"] = df_left_all_rows["Participant"]
df
```

```python
vergence = df_vergence_all_rows[['Veregence Angles Mean [rad]', 'Veregence Angles Std [rad]',
       'Pupil Distance Mean [px]', 'Pupil Distance Std [px]']]
df_all_together = pd.concat([df, vergence], axis=1)
df_all_together

# save in a file
df_all_together.to_csv(r"W:\WCT\04_Mind-Wandering-Labstudy\04_Daten\04_Prepared_data\00_Julia\data_final.csv", index = False)
```
